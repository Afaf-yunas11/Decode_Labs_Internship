```python
import pandas as pd

# -------------------------
# Load Dataset
# -------------------------

df = pd.read_csv("raw_skills.csv")

print("=" * 50)
print("      AI TECH STACK RECOMMENDER")
print("=" * 50)

# -------------------------
# User Input
# -------------------------

user_input = input(
    "\nEnter your skills separated by commas:\n"
)

user_skills = [
    skill.strip().lower()
    for skill in user_input.split(",")
]

# -------------------------
# Recommendation Logic
# -------------------------

recommendations = []

for index, row in df.iterrows():

    job = row["Job Role"]

    job_skills = [
        skill.strip().lower()
        for skill in row["Skills"].split(",")
    ]

    matched = []

    missing = []

    for skill in job_skills:

        if skill in user_skills:
            matched.append(skill)

        else:
            missing.append(skill)

    score = len(matched)

    recommendations.append({

        "Job": job,

        "Score": score,

        "Matched": matched,

        "Missing": missing

    })

# -------------------------
# Sort Results
# -------------------------

recommendations = sorted(

    recommendations,

    key=lambda x: x["Score"],

    reverse=True

)

# -------------------------
# Display Results
# -------------------------

print("\n")
print("=" * 60)
print("TOP 3 RECOMMENDATIONS")
print("=" * 60)

for rec in recommendations[:3]:

    print("\nJob Role :", rec["Job"])

    print("Match Score :", rec["Score"])

    print("Matched Skills :")

    if rec["Matched"]:

        for skill in rec["Matched"]:
            print("   ✔", skill)

    else:
        print("   None")

    print("Missing Skills :")

    for skill in rec["Missing"]:
        print("   ✖", skill)

    print("-" * 60)
```
