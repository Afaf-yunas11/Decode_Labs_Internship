```python
# The Knowledge Base (Dictionary)
responses = {
    'hi': "Hi Afaf, Good Morning! Hope you have a pleasant morning.",
    'what is yap': "YAP is Pakistan's family wallet and SBP-licensed EMI (Electronic Money Institution).",
    'where can i access yap': "You can access YAP via their official website or download the mobile app.",
    'playstore link': "Here is the Play Store link: https://play.google.com/store/apps/details?id=com.yappakistan.app&hl=en-US&pli=1",
    'what is different about it': "YAP is a family super-app with free accounts, affordable cards, and financial literacy tools for all ages.",
    'how do i open an account': "Download the app, enter your details, and verify your identity with your CNIC. No branch visit is needed!",
    'tell me about the young account': "YAP Young is a dedicated app for teenagers (under 18s), which is managed and monitored by parents.",
    'what is different from parents': "Parents hold the main account with full financial control, while children use the separate YAP Young app under parental oversight."
}

# The Main Chatbot Loop
while True:
    # 1. Get input and sanitize it (lowercase + remove extra spaces)
    user_input = input("Ask-Me About YAP: ").lower().strip()

    # 2. Check for the Exit command
    if user_input == 'quit':
        print("Goodbye! Thanks for using YAP Bot.")
        break

    # 3. Look up the reply using the professional .get() method
    reply = responses.get(user_input, 'I do not understand your question about YAP.')
    print(reply)
```
